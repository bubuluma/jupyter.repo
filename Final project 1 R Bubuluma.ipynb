A. A car with a full tank of gas drives 80 miles per hour for 3 hours before running out of gas. How many miles can the car go per tank?


The number of miles the car can go per tank in 3 hours using mathematically Basics.

```python
80*3
```

**Bonus using python**

```python
from matplotlib import pyplot as plt
```

```python
xarray = [1, 3]
yarray = []
for x in xarray:
    y = x * 80
    yarray.append(y)
    print(xarray, yarray)
```

> The answer using python is 240 

 <a id="anchor"></a>
In order to visualise the data Matplotlib package was used 

```python
%matplotlib inLine
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(xarray, yarray)
ax.set_title('Hours vs Miles Graph')
ax.set_ylabel('Hours')
ax.set_xlabel('Miles')
```

>The answer from the graph is 240 miles same as the one from standard mathematically basics

B. A runner travels a half marathon (13.1 miles) in 3 hours. How fast was this runner going in miles per hour?

Calculating using basic mathematics

```python
13.1*3
```

This will be calculated by python

*Firstly the necessary python graphing package was imported*  

```python
from matplotlib import pyplot as plt
```

Next we calculate the number of hours the car travels in 3 hours.

```python
xarray = [1, 3]
yarray = []
for x in xarray:
    y = x * 13.1
    yarray.append(y)
    print(xarray, yarray)
```

>The answer is 39.3 miles per hour

C. The spread of a virus is exponential. If 5 people are infected and each person infects 2 other people, how many people have been infected after 5 rounds of infection?

Formula for exponential infections is y=ab^x where a=initial number of infected virus, b=number infected by one infected virus, and x=number of infection rounds

```python
5*(2**5)
```
